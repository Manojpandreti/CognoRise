Transform data to have a mean of 0 and a standard deviation of 1.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['tickets_sold'] = scaler.fit_transform(data[['tickets_sold']])
```

Rescale the data to a range of [0, 1]

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
data['ticket_price'] = min_max_scaler.fit_transform(data[['ticket_price']])
```

Convert Columns to Correct Data Types

```python
data['date'] = pd.to_datetime(data['date'])
data['cinema_code'] = data['cinema_code'].astype('category')
```

Create new features from the date column like day of the week, month, quarter, or even holidays to better capture patterns.

```python
data['day_of_week'] = data['date'].dt.day_name()
data['is_weekend'] = data['day_of_week'].isin(['Saturday', 'Sunday']).astype(int)
```

The first step in time series analysis is to visualize your data to identify trends, seasonality, and irregularities. Plotting the data helps to see how ticket sales change over time.

```python
import matplotlib.pyplot as plt

# Plot the time series of ticket sales
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['tickets_sold'], color='blue')
plt.title('Time Series Plot of Ticket Sales')
plt.xlabel('Date')
plt.ylabel('Tickets Sold')
plt.show()
```

Decomposition helps break down the time series into its components: Trend, Seasonality, and Residuals (Noise).

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series data
decomposition = seasonal_decompose(data['tickets_sold'], model='additive', period=30)
decomposition.plot()
plt.show()
```

You can use the Augmented Dickey-Fuller (ADF) test to check for stationarity.

```python
from statsmodels.tsa.stattools import adfuller

# Perform the Augmented Dickey-Fuller test
result = adfuller(data['tickets_sold'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
```

Differencing involves subtracting the previous observation from the current observation to remove trends.

```python
data['differenced'] = data['tickets_sold'] - data['tickets_sold'].shift(1)
plt.plot(data['differenced'])
plt.title('Differenced Time Series')
plt.show()
```

Plot Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) to understand the time dependencies.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(data['tickets_sold'].dropna(), lags=30)
plt.show()

plot_pacf(data['tickets_sold'].dropna(), lags=30)
plt.show()
```

Here’s how to use ARIMA (Auto-Regressive Integrated Moving Average), one of the most common time-series forecasting models:

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model to the ticket sales data
model = ARIMA(data['tickets_sold'], order=(1, 1, 1))
model_fit = model.fit()

# Make forecast for the next 10 time periods
forecast = model_fit.forecast(steps=10)
print(forecast)
```

It’s essential to evaluate how well your model is performing on historical data before using it for future predictions.
Common metrics to evaluate forecasting models include:
Mean Absolute Error (MAE):
Mean Squared Error (MSE):
Root Mean Squared Error (RMSE):

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate mean squared error and root mean squared error
mse = mean_squared_error(data['tickets_sold'][-10:], forecast)
rmse = np.sqrt(mse)
print(f'MSE: {mse}, RMSE: {rmse}')
```
